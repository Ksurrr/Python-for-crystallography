# Циклы в Python

В Python существует две основных разновидности циклов:

- циклы, повторяющиеся определенное количество раз (**for**, счетные циклы, counting loops);

- циклы, повторяющиеся до наступления определенного события (**while**, условные циклы, conditional loops).

Цикл for замечательно работает, если мы заранее знаем, сколько повторений (итераций) нам требуется сделать.

## Цикл for

Цикл <b>for</b> исполняется до тех пор, пока условие верно. В Python это записывается следующим образом:

```python
import numpy as np
massiv = np.arange(1,3, 0.5)

for i in massiv: 
    print(i)
```

Двоеточие (:) в конце строки с инструкцией for сообщает интерпретатору Python, что дальше находится блок команд. 

В блок команд входят все строки, расположенные с отступом от строки с инструкцией **for**, вплоть до следующей строки без отступа.

Для повторения цикла некоторое заданное число раз можно использовать цикл <b>for</b> вместе с функцией <b>range</b>.

```python
for i in range(5):
    print('Привет', i)
```

Функция <b>range(n)</b> генерирует последовательность чисел от 0 до n-1, а цикл <b>for</b> последовательно перебирает эту последовательность.

<b>range(n, m)</b>: создает последовательность чисел n, n + 1, n + 2, ..., m - 2, m - 1.

```python
for i in range(2, 5):
    print('Привет', i)
```

## Цикл while

Цикл <b>while</b> – пока условие выполняется, цикл выполняет действие. Применяется когда нужно, чтобы цикл выполнялся до наступления некоторого события, и количество итераций в этом случае заранее оценить просто невозможно.

```python
i = 0
while i < 5:
    print('Привет')
    i += 1
```

Двоеточие (:) в конце строки с инструкцией while сообщает Python, что дальше находится блок команд. В блок входят все строки, расположенные с отступом от строки с инструкцией while, вплоть до следующей строки без отступа.

## Частые сценарии

### Подсчет суммы

Подсчет суммы состоит из двух шагов:

- cоздание переменной сумматора и придание ей первоначального значения: total = 0;

- yвеличение переменной сумматора на нужное число: total = total + num.

```python
# Сложение в цикле
result = 0
for i in range(5):
    result += 2 * (i + 1)
print(result)    
```

### Подсчет произведения

Аналогично сумме вычисляется произведение. При вычислении произведения, начальное значение переменной мультипликатора мы устанавливаем равным 1, в отличии от сумматора, где оно равно 0.

```python
# Умножение в цикле
result = 1
for i in range(5):
    result *= 2 * (i + 1)
print(result)  
```

## Вложенные циклы в Python

Вложенный цикл расположен в еще одном цикле. 

```python
stroki = 4
stolbcy = 6

for i in range(stroki):
    for j in range(stolbcy):
        print('*-', end='')
    print()                 # переход на новую строку
```

# Трехмерные графики

Чтобы построить трехмерный храфик необходимо задать значения z в виде функции от x и y, где для каждой пары x, y есть одно значение z. 

<img src="surface.png"  width=400>

Для этого координаты x, y удобно задавать в виде сетки, воспользовавшись функцией <b>mgrid</b>:

```python
np.mgrid[0:5,0:5]
```

По умолчанию шаг равен 1. Чтобы в <b>mgrid</b> задать начальное значение, конечное значение и шаг, пишем:

```python
np.mgrid[0:1:5j, 0:1:5j]
```

Функция <b>plot_surface</b> строит трёхмерные объекты в виде поверхности. С помощью функции <b>plot_wireframe</b> можно строить трёхмерные сетки. Построение сетки занимает меньше времени и часто используется при визуализации промежуточных расчетов. Для примера построим график функции $z = x * e^{- x^2 - y^2}$

```python
x, y = np.mgrid[-2 : 2 : 40j, -4 : 4 : 20j]
z = x * np.exp(-x**2 - y**2)

fig = plt.figure()                        # создали графическое окно
ax = fig.add_subplot(projection = '3d')   # поместили на него трехмерные координатные оси
ax.plot_wireframe(x, y, z)                # построили график
```

```python
x, y = np.mgrid[-2 : 2 : 40j, -4 : 4 : 20j]
z = x * np.exp(-x**2 - y**2)

fig = plt.figure(figsize=(10, 4))

ax = fig.add_subplot(121, projection='3d')
ax.plot_wireframe(x, y, z)                   # строим сетку

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno')     # строим поверхность
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 6  # число атомов 
a = 2  #  период решетки

x, y = np.mgrid[-0.01 : n*a : 50j, -0.01 : n*a : 50j]   # область определения
V = np.zeros(x.shape)                                   # нулевая матрица для значений потенциала 

fig = plt.figure()                       
ax = fig.add_subplot(projection = '3d')  
ax.plot_surface(x, y, -V, cmap = 'inferno')
ax.set_zlim(-20, 0)
```
