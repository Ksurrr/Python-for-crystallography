Постройте график потенциала $V(x)$ одномерного кристалла, если период решётки $a = 2$, а величина потенциала пропорциональна $Z/x$, где $Z$ – заряд атома, $x$ – расстояние от атома. Пусть $Z = 1$, число атомов на рисунке $n = 6$.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 6 # число атомов 
a = 2 # период решетки

x = np.linspace(0.01, n*a, 101) # область определения
V = np.zeros(len(x))            # нулевой вектор для значений потенциала 

for i in range(n):
    V += np.abs(1/(x - i*a))
    plt.plot(i*a, 0, 'or')
    str1 = plt.text(i*a, 0, '+Z', fontsize=14) 

plt.plot(x, V)
plt.xlabel('r')
plt.ylabel('V(r)')
plt.ylim(-1, 12)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 6  # число атомов 
a = 2  #  период решетки

x, y = np.mgrid[-0.01 : n*a : 50j, -0.01 : n*a : 50j]   # область определения
V = np.zeros(x.shape)                                   # нулевая матрица для значений потенциала 

for i in range(n):
    for j in range(n):
        V +=  np.abs(1/(np.sqrt((x - i*a)**2 + (y - j*a)**2)))
        
fig = plt.figure()                        # создали графическое окно
ax = fig.add_subplot(projection = '3d')   # поместили на него трехмерные координатные оси
ax.plot_surface(x, y, - V, cmap = 'inferno')                # построили график
ax.set_zlim(-20, 0)
```
