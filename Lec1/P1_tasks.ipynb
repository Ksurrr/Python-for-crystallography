# Задача 2

Векторы прямой и обратной решетки кристалла связаны соотношениями:

$ a^* = 2\pi \frac{\mathbf{b}\times \mathbf{c}}{\mathbf{a}\cdot(\mathbf{b}\times \mathbf{c})}, b^* = 2\pi \frac{\mathbf{c}\times \mathbf{a}}{\mathbf{b}\cdot(\mathbf{c}\times \mathbf{a})}, c^* = 2\pi \frac{\mathbf{a}\times \mathbf{b}}{\mathbf{c}\cdot(\mathbf{a}\times \mathbf{b})}.   $ 

Вычислите вектора обратной решетки, если известны вектора прямой.

```python
import numpy as np
a = [2, 0, 0]
b = [0.75, 1.3, 0]
c = [1, 0.25, 1]

a1 = 2 * np.pi * (np.cross(b, c))/(np.dot(a, np.cross(b, c)))
b1 = 2 * np.pi * (np.cross(c, a))/(np.dot(b, np.cross(c, a)))
c1 = 2 * np.pi * (np.cross(a, b))/(np.dot(c, np.cross(a, b)))
print('a1 = ', a1)
print('b1 = ', b1)
print('c1 = ', c1)
```

# Задача 3

Напишите код, который создаёт графическое окно и строит в нем четыре графика – арксинус, арккосинус, гиперболические синус, гипреболический косинус,  на интервале от -1 до 1.

```python
import matplotlib.pyplot as plt 
import numpy as np

# Задаем область определения
x = np.linspace(-1, 1, 99)

# Создаем графическое окно заданного размера:
figure = plt.figure(figsize=(8, 8)) 

# Добавляем на графическое окно координатные оси
ax1 = figure.add_subplot(2, 2, 1) # матрица из графиков 2х2 и номер графика на ней
ax2 = figure.add_subplot(2, 2, 2)
ax3 = figure.add_subplot(2, 2, 3)
ax4 = figure.add_subplot(2, 2, 4)

# В созданных осях строим графики
ax1.plot(x, np.arcsin(x))
ax2.plot(x, np.arccos(x), 'orange')
ax3.plot(x, np.sinh(x), 'green')
ax4.plot(x, np.cosh(x), 'red')

# Подписи по осям
ax3.set_ylabel('y', fontsize=14)
ax1.set_ylabel('y', fontsize=14)
ax3.set_xlabel('x', fontsize=14)
ax4.set_xlabel('x', fontsize=14)

# заголовки
ax1.set_title('Arcsin', fontsize=14)
ax2.set_title('Arccos', fontsize=14)
ax3.set_title('Hyperbolic sin', fontsize=14)
ax4.set_title('Hyperbolic cos', fontsize=14)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
n = 6  # число атомов 
a = 2  #  период решетки
 
x, y = np.mgrid[-0.01 : n*a : 50j, -0.01 : n*a : 50j]   # область определения
V = np.zeros(x.shape)                                   # нулевая матрица для значений потенциала 
 
for i in range(n):
    for j in range(n):
        V += np.abs(1/np.sqrt((x-i*a)**2 + (y-j*a)**2))
        
 
fig = plt.figure()                       
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(x, y, -V, cmap = 'inferno')
ax.set_zlim(-20, 0)
```
