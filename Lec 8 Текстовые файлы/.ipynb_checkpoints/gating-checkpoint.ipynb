```python
raw = "../raw"
```

```python
# 500 ns
scannos = {'180':  15364,
           '90':   15365,
           '0':    15366,
           }
```

```python
# 1000 ns
scannos = {'180':  15348,
           '150':  15349,
           '120':  15350,
           '90':   15351,
           '60':   15352,
           '30':   15353,
           '0':    15354,
           '-30':  15355,
           '-60':  15356,
           '-90':  15357,
           '-120': 15358,
           '-150': 15359,
           '-180': 15360,
           }
```

```python
cmap = cm.get_cmap('jet')

fig, ax = plt.subplots(1,1, figsize=(6,4))

for ii, key in enumerate(scannos):
    fiofile = glob.glob(os.path.join(raw, "*%05i.fio"%scannos[key]))[0]
    fio = rexs.io.FIOdata(fiofile)
    
    omega = fio['omega'] - fio['omega'][0]
    intensity = fio['lmbd_countsroi2_atten'] / fio['eh_c01']
    
    fiofile = glob.glob(os.path.join(raw, "*%05i.fio"%scannos[key]))[0]

    clr = cmap((180 - int(key)) / 360.)
    
    plt.plot(omega, intensity, color=clr, label='%s $^\circ$'%key)

plt.xlabel('$\omega$ ($^\circ$)')
plt.ylabel('Intensity (arb. units)')
plt.legend()
plt.tight_layout()
plt.show()
```
