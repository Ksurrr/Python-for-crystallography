# Настройка графиков в Matplotlib

Все <b>Графики(axes)</b>, которые мы строим в matplotlib, "живут" внутри объекта <b>Рисунок (Figure)</b>. 

То есть <b>Figure (Рисунок)</b> является общим окном, в котором можно построить 1 или несколько <b>графиков (axes)</b>.

<b>Axes</b>- Это сам график. У которого есть оси, заголовок, легенда и т.д.

Если провести аналогию с построением графиков в тетради, то **Figure** - это ваш лист. **Axes** - те области листа, которые вы выделяете под графики. Figure хороша тем, что в ней есть параметры размера картинки. Если вы строите графики, не задав предварительно размер листа, они создаются в некотором размере по умолчанию. При оформлении графиков для научных статей или дипломной работы вам скорее всего захочется задать определенный размер. Но для наших сегодняшних целей вполне хватает размеров по умолчанию. Мы будем обсуждать детальную настройку графиков.

## Параметры функции plot

Самый простой способ построить график, дать функции **plot** в качестве аргумента список значений. В таком случае, plot самостоятельно задаст в качестве значений по оси x числа от 1 и далее с шагом 1, а значения по оси y возьмет из нашего списка:

```python
import matplotlib.pyplot as plt

plt.plot([1, 7, 3, 5, 11, 1])
```

Если мы подаем в качестве аргументов функции **plot** два списка, она использует их как значения по x и y соответственно:

```python
import matplotlib.pyplot as plt

plt.plot([1, 5, 10, 15, 20], [1, 7, 3, 5, 11])
plt.show()
```

Для добавления на график подписей по осям, заголовка и сетки, используем соответствующие функции:

```python
import numpy as np

x = np.linspace(1, 50, 100)
y = 3 * x - 2

plt.title("Линейная зависимость y = 3x - 2")  # заголовок
plt.xlabel("x")  # ось абсцисс
plt.ylabel("y")  # ось ординат
plt.grid()  # включение отображение сетки
plt.plot(x, y)  # построение графика

plt.show()
```

## Цвет и толщина линии, точки на графике

Изменим тип линии и ее цвет, для этого в функцию plot(), в качестве третьего параметра передадим строку, сформированную определенным образом: **'r–-'**

Такой строкой можно закодировать цвет, вид точек и тип линии, собрав её их следующих элементов:

             b     blue          .     point              -     solid
             g     green         o     circle             :     dotted
             r     red           x     x-mark             -.    dashdot 
             c     cyan          +     plus               --    dashed   
             m     magenta       *     star             (none)  no line
             y     yellow        s     square
             k     black         d     diamond
             w     white         v     triangle (down)
                                 ^     triangle (up)
                                 <     triangle (left)
                                 >     triangle (right)
                                 p     pentagram
                                 h     hexagram

Чтобы изменить толщину линии, пользуемся параметром **linewidth**, которому можно присваивать численные значения:

```python
import numpy as np

x = np.linspace(1, 50, 100)
y = 3 * x - 2

plt.title("Линейная зависимость y = 3x - 2")  # заголовок
plt.xlabel("x")  # ось абсцисс
plt.ylabel("y")  # ось ординат
plt.grid()  # включение отображение сетки
plt.plot(x, y, "r--", linewidth=3)  # построение графика

plt.show()
```

Все типы линий, маркеров и основные цвета можно найти в документации: https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

Другой способ задать вид линии, прописать цвет и маркер в виде аргументов функции plot с заданными значениями. Такой код займёт больше места, но будет проще читаться:

```python
import matplotlib.pyplot as plt

plt.plot([1, 5, 10, 15, 20], [1, 7, 3, 5, 11], color="r", marker="^")
plt.show()
```

То же самое в краткой записи:

```python
import matplotlib.pyplot as plt

plt.plot([1, 5, 10, 15, 20], [1, 7, 3, 5, 11], "r-^")
plt.show()
```

## Параметры функций xlabel, ylabel

Документация по управлению текстом на графиках: https://matplotlib.org/stable/tutorials/text/text_intro.html#

Основные параметры:

- текст подписи **xlabel**, str;
- расстояние от осей до подписи **labelpad**, число (по умолчанию None);
- размер шрифта **fontsize** или **size**, число или значение из {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'};
- стиль шрифта **fontstyle**, значение {'normal', 'italic', 'oblique'};
- толщина шрифта **fontweight**, число в диапазоне от 0 до 1000 либо значение из {'ultralight', 'light', 'normal', 'regular', 'book', 'medium', 'roman', 'semibold', 'demibold', 'demi', 'bold', 'heavy', 'extra bold', 'black'} 
- цвет шрифта **color**, значение из набора  {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}, либо другие способы задания цвета https://matplotlib.org/stable/tutorials/colors/colors.html
- для изменения шрифта на Times New Roman перед графиком добавляем специальную строку 


```python
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times New Roman"

plt.plot([1, 5, 10, 15, 20], [1, 7, 3, 5, 11])
plt.xlabel("x", labelpad=10, fontsize=18, fontstyle="italic")
plt.ylabel("y", fontweight="heavy", color="g", fontsize=18)
plt.title("Заголовок", fontsize=18)
plt.show()
```

## Параметры функции title



- Текст заголовка, строка;
- положение заголовка **loc**: значение из набора {'center', 'left', 'right'};
- все параметры, касающиеся оформления текста, такие же как у xlabel, ylavel.


```python
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
abs_x = [3, 2, 1, 0, 1, 2, 3]

# стиль графика
# plt.style.use('default')
plt.rcParams["font.family"] = "Arial"

plt.plot(x, y)
plt.plot(x, abs_x)

# подписи по осям
plt.xlabel("x")
plt.ylabel("y")

# заголовок
plt.title("Заголовок", loc="left", fontsize=18, fontweight="extra bold")
```

## Легенда

Легенда будет размещена на графике, если вызвать функцию **legend()**, параметры - названия кривых на графике в виде списка.

```python
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
abs_x = [3, 2, 1, 0, 1, 2, 3]

plt.plot(x, y)
plt.plot(x, abs_x)

# подписи по осям
plt.xlabel("x")
plt.ylabel("y")

# легенда
plt.legend(["$y = x^2$", "$y = |x|$"])

# заголовок
plt.title("Заголовок", loc="left", fontsize=18, fontweight="extra bold")
```

Названия графиков можно указывать в момент построения, используя параметр **label**, а потом вызывать функцию **legend** без параметров.

```python
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
abs_x = [3, 2, 1, 0, 1, 2, 3]

plt.plot(x, y, label="$y = x^2$")
plt.plot(x, abs_x, label="$y = |x|$")

# подписи по осям
plt.xlabel("x")
plt.ylabel("y")

# легенда
plt.legend()

# заголовок
plt.title("Заголовок", fontsize=18)
```

## Текстовое примечание

За размещение текста на поле графика отвечает функция **text()**, которой вначале передаются координаты позиции надписи, после этого – текст самой надписи

```python
x = np.linspace(0, 6, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Задаем единый размер шрифта для всех элементов графика
plt.rcParams.update({"font.size": 16})

plt.plot(x, y1, "r")
plt.plot(x, y2, "b")
plt.xlabel("x")
plt.ylabel("y")

plt.text(2.5, 0.75, "sin(x)", color="r")
plt.text(0.7, -0.25, "cos(x)", color="b")
```

## Стили графиков

Разные варианты оформления графиков, которые уже придуманы до нас. Подключаются с помощью функции  **plt.style.use()**. Посмотреть все варианты можно по ссылке https://www.dunderdata.com/blog/view-all-available-matplotlib-styles, или используюя следующий код:

```python
x = np.linspace(-3, 3, 30)
y = x ** 2

fig = plt.figure(dpi=100, figsize=(10, 20), tight_layout=True)
available = ["default"] + plt.style.available
for i, style in enumerate(available):
    with plt.style.context(style):
        ax = fig.add_subplot(10, 3, i + 1)
        ax.plot(x, y)
    ax.set_title(style)
```

## Графики, заданные параметрически

Графики, заданные в параметрической форме $x = f(t)$, $y = g(t)$, строятся по следующему алгоритму: 

1. определить массив t;

2. определить массивы $x = f(t)$, $y = g(t)$;

3. построить график plot(x,y).

Для примера построим окружность заданную параметрически:

$\begin{cases} 
x = x_0 + R cos(\phi), 
\\ y = y_0 + R sin(\phi). \end{cases}$


```python
import matplotlib.pyplot as plt
import numpy as np

# Координаты центра окружности
x0 = 1
y0 = 1

Radius = 1  # радиус окружности

# Задаем область определения параметра
phi = np.arange(0, 2.1 * np.pi, np.pi / 50)

x = x0 + Radius * np.cos(phi)
y = y0 + Radius * np.sin(phi)

plt.plot(x, y)
plt.axis("equal")  # задаем равные единичные отрезки по осям
plt.title("Окружность")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## Двойные координатные оси

Для визуализации математических функций с сильно отличающимися значениями существуют функции **ax.twinx()** для оси OX и **ax.twiny** для оси OY, которые выводят две координатные оси с разными масштабами. 


Построим на одном графике зависимости $f = \sin(x)$ и $g = 10 \sin(x)$.

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 10, 0.1)
f = np.sin(3 * t)
g = 10 * np.cos(t)

fig, ax1 = plt.subplots()

ax1.set_xlabel("t")
ax1.set_ylabel("$f = \sin(x)$", color="b")
ax1.plot(t, f, color="b")
ax1.tick_params(axis="y", labelcolor="b")

ax2 = ax1.twinx()

ax2.set_ylabel("$g = 10 \sin(x)$", color="orange")
ax2.plot(t, g, color="orange")
ax2.tick_params(axis="y", labelcolor="orange")

fig.suptitle("Axes.twinx() function Example", fontweight="bold")

plt.show()
```

## Графики в полярных координатах

Для создания полярных координат, задаем координатам свойство 'polar': **plt.axes(polar=True)}** или выбираем проекцию для графика **subplot_kw='projection': 'polar'**.

Построим график функции $R = | sin(\phi) cos(\phi) |$.

```python
import matplotlib.pyplot as plt
import numpy as np

fi = np.arange(0, 2.1 * np.pi, np.pi / 40)
Ro = np.abs(np.sin(fi) * np.cos(fi))

fig, ax = plt.subplots(subplot_kw={"projection": "polar"})
ax.plot(fi, Ro)
ax.set_rticks([0.2, 0.4])  # Less radial ticks
ax.grid(True)

ax.set_title("$R = | sin(\phi) cos(\phi) |$", va="bottom")
plt.show()
```
