Пользуясь документацией, постройте на одном графике функции 

 $ y = \cos (2 \pi x) exp (-x)$ и $y = \cos(2 \pi x) $,

х задан в диапазоне от 0 до 5.
Первый график должен быть жирной красной линией, второй синими точками.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y1 = np.cos(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x) * np.exp( -x )

plt.plot(x, y2, 'r-', linewidth = 3)
plt.plot(x, y1, 'b:')
```
